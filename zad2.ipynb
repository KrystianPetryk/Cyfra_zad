```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image


im = np.array(Image.open('zad2_1.jpg')).mean(axis=2)/255
im2= np.array(Image.open('zad2_2.jpg')).mean(axis=2)/255

def draw_image(im):
  plt.figure(figsize=(12,9))
  plt.imshow(im, cmap='gray')
  plt.clim(0, 1)
  plt.colorbar()
  plt.show()

def draw_image_white(im):
  plt.figure(figsize=(12,9))
  plt.imshow(im*2, cmap='gray')
  plt.clim(0, 1)
  plt.colorbar()
  plt.show()

def draw_image_gamma(im):
  plt.figure(figsize=(12,9))
  plt.imshow(im**0.4, cmap='gray')
  plt.clim(0, 1)
  plt.colorbar()
  plt.show()
```

```python
 

draw_image(im)
```

```python
draw_image_white(im)
```

```python
draw_image(im2)
```

```python
draw_image_gamma(im2)
```
